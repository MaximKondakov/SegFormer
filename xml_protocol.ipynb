```python
def create_region_property(region, count_region, label_xy_list):
    ET.SubElement(region, 'display').text = 'true'   
    ET.SubElement(region, 'line_color').text = f'{get_line_color()}'
    ET.SubElement(region, 'line_visibility').text = f'{get_line_visibility()}'
    ET.SubElement(region, 'line_width').text = f'{get_line_width()}'
    ET.SubElement(region, 'line_type').text = f'{get_line_type()}'
    ET.SubElement(region, 'fill_color').text = f'{get_fill_color()}'
    ET.SubElement(region, 'fill_visibility').text = f'{get_fill_visibility()}'
    ET.SubElement(region, 'label_text').text = f'Область {count_region}'
    ET.SubElement(region, 'label_x').text = f'{label_xy_list[i_region][0]}'
    ET.SubElement(region, 'label_y').text = f'{label_xy_list[i_region][1]}'    
    ET.SubElement(region, 'label_color').text = f'{get_label_color()}'
    ET.SubElement(region, 'label_visibility').text = f'{get_label_visibility()}'
    ET.SubElement(region, 'label_font_size').text = f'{get_label_font_size()}'
    ET.SubElement(region, 'label_font_bold').text = f'{get_label_font_bold()}'
    ET.SubElement(region, 'label_font_italics').text = f'{get_label_font_italics()}'
    ET.SubElement(region, 'label_font_underline').text = f'{get_label_font_underline()}'
    ET.SubElement(region, 'region_hint').text = f'Область {count_region}'
    ET.SubElement(region, 'region_z').text = f'{get_region_z()}'
    ET.SubElement(region, 'region_weight').text = f'{get_region_weight()}'
        
```

```python
def create_region(region, i_region, label_xy_list, regions_points, conf_i_class_list, regions_table_name):
    global count_region
    count_region += 1
    create_region_property(region, count_region, label_xy_list)
    regions_table = ET.SubElement(region, 'regions_table')####regions_table
    ET.SubElement(regions_table, 'column', attrib={'id':'1','value':f'Область {count_region}'},
                  hint=f'Область {count_region}')
    
    if regions_table_name == 'tsd':
        ET.SubElement(regions_table, 'column', attrib={'id':'2',
                                                       'value':f'{conf_i_class_list[i_region]}'})          
        ET.SubElement(regions_table, 'column', attrib={'id':'3','value':f'{get_region_weight()}'})
    elif regions_table_name == 'heterogeneity':
        ET.SubElement(regions_table, 'column', attrib={'id':'2',
                                                       'value':f'{conf_i_class_list[i_region]}'})                  
    elif regions_table_name == 'illegal':
        ET.SubElement(regions_table, 'column', attrib={'id':'2','value':f'{get_region_name()}'},
                     hint = f'{get_region_name()}')
        ET.SubElement(regions_table, 'column', attrib={'id':'3',
                                                       'value':f'{conf_i_class_list[i_region]}'})      
        
    
    points  = ET.SubElement(region, 'points', attrib={'count':f'{len(regions_points[i_region])}'})
    for i in range(len(regions_points[i_region])):
        ET.SubElement(points, 'point', attrib={'id':f'{i+1}',
                                               'x':f'{regions_points[i_region][i][0]}',
                                               'y':f'{regions_points[i_region][i][1]}'
                                              })
```
